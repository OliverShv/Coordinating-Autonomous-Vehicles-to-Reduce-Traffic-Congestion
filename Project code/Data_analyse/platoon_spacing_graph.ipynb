```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(6,32,1000)
y = []

d = 10

y2 = [0.75,1.5]
x2 = [d, d]

for i in x:
    if i >= d:
        b = (i+1-d)**0.135
        if b > 1.5:
            b = 1.5
        y.append(b)
    else:
        b = ((i-1-d)**2)**0.5
        c = 1/(b**(b*0.1))
        print(c)

        if c < 0.75:
            c = 0.75
        y.append(c)
```

```python
plt.figure(figsize=[10,10])
plt.title("Speed multiplier based on the distance between 2 vehicles \n in a platoon with a set distance of "+str(d)+" metres", fontsize=18)
plt.xlim([5, 33])
plt.xlabel('Distance', fontsize=16)
plt.ylabel('Multiplier', fontsize=15)
plt.plot(x,y)
plt.plot(x2,y2)
plt.legend(["Speed Multiplier","Set spacing"])
```
